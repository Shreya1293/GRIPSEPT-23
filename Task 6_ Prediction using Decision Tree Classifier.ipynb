##### Decision Tree: 
A Decision Tree Classifier is a popular machine learning algorithm used for prediction modeling. It works by recursively splitting the dataset into subsets based on the most significant features, forming a tree-like structure. At each node, it selects the feature that best separates the data, resulting in a model that's easy to visualize and interpret. Decision trees are effective for both classification and regression tasks, making them a versatile choice for predictive modeling in various domains.

##### Objective:
The primary goal is to construct a Decision Tree classifier and represent it graphically. This classifier's core purpose is to enable accurate predictions of the correct class when provided with new data. 

The essential tools employed for this task include Pandas, NumPy, Scikit-Learn, and Pydotplus. 

The dataset utilized for this project is the Iris.csv dataset.

**1. Importing the necessary library**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn import metrics 
from sklearn.tree import DecisionTreeClassifier
from matplotlib import pyplot as plt
```

**2. Loading the dataset** 

```python
df = pd.read_csv('Iris.csv')
df.head()
```

```python
df.info()
```

**3. Feature Selection**

```python
X = df.iloc[:,[1,2,3,4]].values
print(X)

y=df.Species
print("The target classes\n", y)
```

**4. Splitting the Datasets**

```python
#splitting the data into 70% and 30%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

**5. Building Decision Tree Classifier**

```python
#using entropy as the attribute selection measure and maximum depth, 4
model = DecisionTreeClassifier(criterion='entropy', splitter='best', max_depth=4)

#training the model
model = model.fit(X_train,y_train)

#predicting the output for test dataset
y_pred= model.predict(X_test)
```

**6. Evaluation of the model**

```python
#Accuracy is calculated by comparing actual test set values and predicted values.
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

The model predicts with accuracy of 95%. We can improve the accuracy by changing the parameters of the model. Here, I'll be chnaging the depth of the decision tree from 4 to 3.

```python
model = DecisionTreeClassifier(criterion='entropy', splitter='best', max_depth=3)

model = model.fit(X_train,y_train)

y_pred= model.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

The accuracy has imporved from 95% to 97%

**7. Visualisation**

```python
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(model, 
                   feature_names=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'],  
                   class_names=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'],
                   filled=True)
```
